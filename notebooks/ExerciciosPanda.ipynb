```python
import pandas as pd
import numpy as np
import math

# Montando o Drive
from google.colab import drive
drive.mount('/content/drive')

# Caminho do diretório escolhido
path = "/content/drive/MyDrive/AnaliseDeDados"

# Leitura do arquivo
df_jogos = pd.read_csv(path + '/vgsales.csv', sep=',')

df_jogos
```

```python
soma_vendas_por_genero = df_jogos.groupby('Genre')['Global_Sales'].sum()
soma_vendas_por_genero

genero_com_maior_venda = soma_vendas_por_genero.idxmax()
vendas_do_genero = soma_vendas_por_genero.max()

print("Gênero com maior numero de vendas foi", genero_com_maior_venda, "com um total de vendas de", vendas_do_genero)
```

```python
soma_vendas_por_plataforma = df_jogos.groupby('Platform', as_index=False)['NA_Sales'].mean()
top_5 = soma_vendas_por_plataforma.sort_values(by='NA_Sales', ascending=False).head(5)
print(top_5)
```

```python
filtro_periodo = df_jogos[(df_jogos['Year'] >= 2000) & (df_jogos['Year'] <= 2010)]

contagem_por_publisher = filtro_periodo['Publisher'].value_counts()

publisher_top1 = contagem_por_publisher.idxmax()
quantidade_top1 = contagem_por_publisher.max()

print(f"O publisher que lançou mais jogos entre 2000 e 2010: {publisher_top1} com {quantidade_top1} jogos")
```

```python
media_vendas_jp_por_genero = df_jogos.groupby('Genre')['JP_Sales'].mean().sort_values(ascending=False)
media_vendas_jp_por_genero
```

```python
mediana_vendas_jp_por_genero = df_jogos.groupby('Genre')['JP_Sales'].median().sort_values(ascending=False)
mediana_vendas_jp_por_genero
```

Analisando as vendas no Japão (JP_Sales) por gênero, observamos que o gênero Role-Playing lidera tanto na média quanto na mediana, o que confirma sua forte popularidade no mercado japonês.

Fighting também aparece bem posicionado em ambas as métricas, indicando boa aceitação.

Porém, há uma diferença marcante no caso de Adventure e Platform:

Platform aparece no top 3 pela média, mas não pela mediana — o que pode indicar a presença de alguns poucos jogos com vendas muito altas, que puxam a média para cima.

Já Adventure, que está entre os últimos na média, aparece como segundo lugar na mediana — sugerindo que a maioria dos jogos nesse gênero tem vendas modestas, mas consistentes, sem grandes outliers.
